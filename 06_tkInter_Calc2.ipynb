Creating more use out of the calculator!

```python
from tkinter import *

root = Tk()
root.title("More Advanced Calculator")

# entry box widget
e= Entry(root, width=35, borderwidth=5)
e.grid(row=0,column=0, columnspan=3, padx=10, pady=10)

n = 0
y = 0

def button_click(number):
    
    current = e.get()
    e.delete(0, END)
    e.insert(0, str(current) + str(number))
    
def button_clear():
    global n
    global y
    n=0
    y=0

    e.delete(0,END)
    
def button_add():
    
    global add1
    global n
    
    if(n != 0 or y != 0):
        button_equal()
        button_add()
        return
    elif(n == 0):
        add1 = e.get()
        e.delete(0,END)
        n+=1

def button_equal():
    
    global n
    global y
    global total
    
    num2 = e.get()
    e.delete(0,END)
    
    if(n > 0):
        total = float(add1) + float(num2)
        n=0
        e.insert(0, total)
        return
    if(n < 0):
        total = float(sub1) - float(num2)
        n=0
        e.insert(0, total)
        return
    if(y > 0):
        total = float(mult1) * float(num2)
        y=0
        e.insert(0,total)
        return
    if(y < 0):
        total = float(div1) / float(num2)
        y=0
        e.insert(0,total)
        return

def button_sub():
    global sub1
    global n
    if(n != 0 or y != 0):
        button_equal()
        button_sub()
        return
    else:
        sub1 = e.get()
        e.delete(0,END)
        n-=1
    
def button_mult():
    global mult1
    global y
    if(n != 0 or y != 0):
        button_equal()
        button_mult()
        return
    else:
        mult1 = e.get()
        e.delete(0,END)
        y+=1
        
def button_div():
    global div1
    global y
    if(n != 0 or y != 0):
        button_equal()
        button_div()
        return
    else:
        div1 = e.get()
        e.delete(0,END)
        y-=1

#Initializing Calc Buttons

b_1 = Button(root, text="1", padx=40, pady=20, command= lambda:button_click(1))
b_2 = Button(root, text="2", padx=40, pady=20, command= lambda:button_click(2))
b_3 = Button(root, text="3", padx=40, pady=20, command= lambda:button_click(3))
b_4 = Button(root, text="4", padx=40, pady=20, command= lambda:button_click(4))
b_5 = Button(root, text="5", padx=40, pady=20, command= lambda:button_click(5))
b_6 = Button(root, text="6", padx=40, pady=20, command= lambda:button_click(6))
b_7 = Button(root, text="7", padx=40, pady=20, command= lambda:button_click(7))
b_8 = Button(root, text="8", padx=40, pady=20, command= lambda:button_click(8))
b_9 = Button(root, text="9", padx=40, pady=20, command= lambda:button_click(9))
b_0 = Button(root, text="0", padx=40, pady=20, command= lambda:button_click(0))
b_dec = Button(root, text=".", padx=40, pady=20, command= lambda:button_click("."))

b_add = Button(root, text="+", padx=39, pady=20, command= button_add)
b_sub = Button(root, text="-", padx=39, pady=20, command= button_sub)
b_mult = Button(root, text="*", padx=39, pady=20, command= button_mult)
b_div = Button(root, text="/", padx=39, pady=20, command= button_div)

b_equal = Button(root, text="=", padx=30, pady=20, command= button_equal)
b_clear = Button(root, text="Clear", padx=30, pady=20, command = button_clear)

#Actual Buttons for Calc

b_1.grid(row=3, column=0)
b_2.grid(row=3, column=1)
b_3.grid(row=3, column=2)

b_4.grid(row=2, column=0)
b_5.grid(row=2, column=1)
b_6.grid(row=2, column=2)

b_7.grid(row=1, column=0)
b_8.grid(row=1, column=1)
b_9.grid(row=1, column=2)

b_0.grid(row=4, column=1)
b_dec.grid(row=4, column=0)
b_add.grid(row=4, column=2)
b_sub.grid(row=5, column=0)
b_mult.grid(row=5, column=1)
b_div.grid(row=5, column=2)

b_equal.grid(row=6, column=1, columnspan=2)
b_clear.grid(row=6, column=0, columnspan=2)



root.mainloop()
```

This calculator can act as a basic 4 operation functioning calculator. It can give you any answer to a multi operation problem. However it doesn't read and react to PEMDAS, you must enter it in the order of pemdas for it to work like that
